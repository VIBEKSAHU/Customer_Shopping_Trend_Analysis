```python
import pandas as pd
df=pd.read_csv("customer_shopping_behavior.csv")
```

```python
#check for missing values
df.isnull().sum()
```

```python
#filling the null values
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

```

```python
df.isnull().sum()
```

```python
#snack casing /to transform the upper case letter into lower case so that errors should not occur
```

```python
df.columns
```

```python
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column age_group for better analysis
```

```python
labels=['Young_Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create a column purchase_frequency_days
```

```python
#created a dictionary first to store the corresponding numerical vale for each text value then to map it .
frequency_mapping={
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine

# Step 1: Connect to PostgreSQL
# Replace placeholders with your actual details
username = "postgres"      # default user
password = "root" # the password you set during installation
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
database = "customer_behavior"    # the database you created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
table_name = "customer"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
